```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 500,n_features = 4,centers = 8,cluster_std = 0.8,random_state = 101)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],data = data,c = data[1],cmap = 'rainbow')
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 8)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,sharey = True,figsize = (10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c = kmeans.labels_,cmap = 'rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c = data[1],cmap = 'rainbow')

```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(data[1],kmeans.labels_))
print(classification_report(data[1],kmeans.labels_))
```

```python
file = r'C:\Users\hp\Desktop\iris.csv'
df = pd.read_csv(file)
```

```python
df.head()
```

```python
df.drop('Id',axis = 1,inplace = True)
```

```python
sns.set_style('whitegrid')
sns.lmplot(x = 'SepalLengthCm',y = 'SepalWidthCm',data=df, hue='Species',
           palette='rainbow',height=6,aspect=1,fit_reg=False)
```

```python
sns.set_style('whitegrid')
sns.lmplot(x = 'SepalWidthCm',y = 'PetalWidthCm',data=df, hue='Species',
           palette='inferno',height=6,aspect=1,fit_reg=False)
```

```python
sns.set_style('whitegrid')
sns.lmplot(x = 'PetalLengthCm',y = 'PetalWidthCm',data=df, hue='Species',
           palette='magma',height=6,aspect=1,fit_reg=False)
```

```python
Kmeans = KMeans(n_clusters = 3)
Kmeans.fit(df.drop('Species',axis = 1))
```

```python
Kmeans.cluster_centers_.shape
```

```python
Kmeans.labels_
```

```python
def converter(species):
    if species=='Iris-setosa':
        return 2
    elif species == 'Iris-versicolor':
            return 1
    else:
            return 0
```

```python
df['Cluster'] = df['Species'].apply(converter)
```

```python
df.info()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],Kmeans.labels_))
print(classification_report(df['Cluster'],Kmeans.labels_))
```
